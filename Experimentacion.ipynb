## Formación del dataset final

```python
import functools as ft
```

```python
# lst_datasets = [universe, transactionality, taxpayer, movements, legal_representatives, balances, acquisition]
lst_datasets = [universe, transactionality, taxpayer, movements, legal_representatives, acquisition]
```

```python
dataset = ft.reduce(lambda left, right: pd.merge(left, right, on=['period', 'ID'], how='left'), lst_datasets)
```

```python
dataset.drop(columns=['ID', 'period'], inplace = True)
```

```python
dataset.head()
```

```python
dataset.describe()
```

## Formación de los subsets

```python
from sklearn.model_selection import train_test_split
```

```python
y= dataset['target']
X= dataset.drop(columns=['target'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Preprocesamiento de los Datos

### Verificacion de valores faltantes

```python
cols = X_train.isna().sum()
print(cols)
```

### Se eliminan las características con muchos valores faltantes

```python
columnas_eliminadas = X_train.columns[X_train.isnull().mean() > 0.5].tolist()
X_train.drop(columns = columnas_eliminadas, inplace = True)
X_train.head()
```

```python
print(columnas_eliminadas)
```

### Imputación de datos

```python
import numpy as np
import statistics as st
from pprint import pprint
```

```python
X_train.columns
```

#### Se utiliza a la media para las características numéricas

```python
lst_caract_num = [
    'time_from_start', 'number_legal_rep', 'specialized_legal_rep', 'income_legal_rep', 'min_age', 'max_age',
    'time_default', 'number_companies', 'number_credit_acquisitions', 'total_amount', 'max_amount', 'min_amount'
]

medias_dict = {}
for columna in lst_caract_num:
    media_caract = X_train[columna].astype("float").mean(axis=0)
    medias_dict[columna] = media_caract
    X_train[columna].replace(np.nan, media_caract, inplace=True)
    
pprint(medias_dict)
```

#### Se utiliza a la moda para las características categóricas

```python
lst_caract_categ = [
    'category', 'activity', 'economic_category', 'road', 'license', 'has_insurance', 'normal_balance', 'bureau_risk'
]

modas_dict = {}
for columna in lst_caract_categ:
    moda_caract = st.mode(X_train[columna])
    modas_dict[columna] = moda_caract
    X_train[columna].replace(np.nan, moda_caract, inplace=True)

pprint(modas_dict)
```
